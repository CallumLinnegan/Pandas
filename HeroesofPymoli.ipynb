## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
PA_age_bins = [5, 9, 14, 19, 24, 29, 34, 39, 46]
PA_groups_names = ["5-10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40 +"]

#Created new bin, previous is unique ID's only
df["Age Group"] = pd.cut(df["Age"], bins=PA_age_bins, labels=PA_groups_names)

#divide dataframe based on age group
age_group = df.groupby("Age Group")

#number of purchases per group
age_purchase_count = age_group["SN"].count()

#mean price per group
mean_price = round(age_group['Price'].mean(),2)

#total purchase value
total_purchase = round(age_group["Price"].sum(),2)

Avg_total = round((age_group["Price"].sum()/total_count_age),2)

Purchasing_analysis = pd.DataFrame({'Purchase Count':age_purchase_count,
                                    'Average Purchase Price':mean_price,
                                    'Total Purchase Value':total_purchase,
                                    'Avg Total Purchase per Person':Avg_total})
Purchasing_analysis["Average Purchase Price"] = "$" + Purchasing_analysis['Average Purchase Price'].astype(str)
Purchasing_analysis["Total Purchase Value"] = "$" + Purchasing_analysis['Total Purchase Value'].astype(str)
Purchasing_analysis["Avg Total Purchase per Person"] = "$" + Purchasing_analysis['Avg Total Purchase per Person'].astype(str)
Purchasing_analysis
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#group by username (SN)
Top_spenders = df.groupby("SN")
Purchase_count = Top_spenders["Purchase ID"].count()
Avg_price = round(Top_spenders["Price"].mean(),2)
total_value = round(Top_spenders["Price"].sum(),2)

TopSpenders = pd.DataFrame({'Purchase Count':Purchase_count,
                            'Average Purchase Price':Avg_price,
                            'Total Purchase Value':total_value})
TopSpenders
#Sort by the total purchase values (highest to lowest)
TopSpenders1 = TopSpenders.sort_values(["Total Purchase Value"],ascending = False)
TopSpenders1["Average Purchase Price"] = "$" + TopSpenders1['Average Purchase Price'].astype(str)
TopSpenders1["Total Purchase Value"] = "$" + TopSpenders1['Total Purchase Value'].astype(str)
TopSpenders1
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
Pop_Items = df[["Item ID","Item Name","Price"]]
Pop_Items1 = Pop_Items.groupby(["Item ID","Item Name"])
Pur_count = Pop_Items1["Price"].count()
Total_price = round(Pop_Items1["Price"].sum(),2)

Item_Price = round(Total_price / Pur_count,2)
Item_Price

Most_Popular = pd.DataFrame({'Purchase count':Pur_count,
                             'Item Price':Item_Price,
                             'Total Purchase Value':Total_price})

Most_Popular["Item Price"] = "$" + Most_Popular['Item Price'].astype(str)
Most_Popular["Total Purchase Value"] = "$" + Most_Popular['Total Purchase Value'].astype(str)
Most_Popular
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order

```python
Most_Profitable = Most_Popular.sort_values(['Total Purchase Value'],ascending = False)
Most_Profitable
```
